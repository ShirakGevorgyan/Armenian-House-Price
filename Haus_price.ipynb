```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('Hous_Price.csv')
```

```python
df.drop(337, inplace=True)
```

```python
df["Գինը"] = df["Գինը"].apply(lambda string: string.lstrip("$") if "պայմ." not in string else np.nan ).astype(float)
```

```python
df.info()
```

```python
df = df[df["Գինը"]>30000]
df.dropna()
```

```python
X = df.drop('Գինը', axis=1)
y = df['Գինը']
```

```python
X
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
for i in X.columns:
    if X[i].dtype == object:
        encoder = LabelEncoder()
        X[i] = encoder.fit_transform(X[i])
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model = RandomForestRegressor()
```

```python
hyper_param = {
    'n_estimators': [100, 150],
    'criterion': ["squared_error", "absolute_error", "friedman_mse", "poisson"],
    'max_depth': [None, 10],
    'min_samples_split': [2, 5],
    'min_samples_leaf': [1, 2],
    'min_weight_fraction_leaf': [0.0, 0.1],
    'max_leaf_nodes': [None, 10],
    'min_impurity_decrease': [0.0, 0.1]
}
```

```python
full_model = GridSearchCV(model, hyper_param, cv=3, verbose=2)
```

```python
full_model.fit(X_train, y_train)
```

```python
y_pred = full_model.predict(X_test)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
mean_absolute_error(y_pred, y_test)
```

```python
np.sqrt(mean_squared_error(y_pred, y_test))
```

```python
r2_score(y_test, y_pred)
```
